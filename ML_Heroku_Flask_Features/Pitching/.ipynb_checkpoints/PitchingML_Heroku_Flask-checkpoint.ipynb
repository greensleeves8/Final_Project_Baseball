```python
import pandas as pd
from pathlib import Path
```

```python
# Import CSV
file_path = Path('hall_pitching.csv')
hall_pitching_df = pd.read_csv(file_path)
hall_pitching_df
```

```python
# Import CSV
file_path = Path('Player_Names.csv')
Player_Names_df = pd.read_csv(file_path)
Player_Names_df
```

```python
Player_Names_df.drop(columns = ['Unnamed: 0'])
```

```python
Names_Pitching_df = Player_Names_df.merge(hall_pitching_df, left_on = 'playerID', right_on = 'playerID')
```

```python
Names_Pitching_df
```

```python
Names_Pitching_df = Names_Pitching_df.drop(columns = ["Unnamed: 0", "playerID", "inducted"])
```

```python
Names_Pitching_df
```

```python
Names_Pitching_df.to_json('Names_Pitching.json', orient = 'records')
```

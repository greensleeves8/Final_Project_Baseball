```python
import pandas as pd
from pathlib import Path
```

```python
# Import CSV
file_path = Path('hall_batter.csv')
hall_batter_df = pd.read_csv(file_path)
hall_batter_df
```

```python
# Import CSV
file_path = Path('Player_Names.csv')
Player_Names_df = pd.read_csv(file_path)
Player_Names_df
```

```python
list(Player_Names_df.columns)
```

```python
Player_Names_df.drop(columns = ['Unnamed: 0'])
```

```python
Names_Batter_df = Player_Names_df.merge(hall_batter_df, left_on = 'playerID', right_on = 'playerID')
```

```python
Names_Batter_df
```

```python
Names_Batter_df = Names_Batter_df.drop(columns = ["Unnamed: 0", "playerID", "inducted"])
```

```python
Names_Batter_df
```

```python
Names_Batter_df.to_json('Names_Batter.json', orient = 'records')
```

```python
import json
Names_Batter = Names_Batter_df.to_json(orient = 'records')
parsed = json.loads(Names_Batter)
json.dumps(parsed, indent = 4)
```

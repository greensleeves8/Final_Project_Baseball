```python
# importing dependencies for Batting.csv data cleaning/ETL

import pandas as pd
from pathlib import Path


file_path = Path('./Resources/Original_CSV/Batting.csv')
batting_df = pd.read_csv(file_path)
batting_df
```

```python
# Inspecting batting_df headers

batting_df.columns
```

```python
# Reading in HallOfFame.csv to merge with batting_df and create target

file_path_HOF = Path('./Resources/Original_CSV/HallOfFame.csv')
HOF_df = pd.read_csv(file_path_HOF)
HOF_df
```

```python
HOF_df = HOF_df.loc[(HOF_df['category'] == 'Player')]
HOF_df
```

```python
# checking for null values

batting_df.isnull().sum()
```

```python
# dropping CS, IBB, GIDP columns due to high rate of null values 
# SF not dropped due to being in SLG % formula

batting_df = batting_df.drop(columns=['CS', 'IBB', 'GIDP']).copy()
batting_df
```

```python
# Creating new dataframe of players' cumulative career statistics

career_batter_df = batting_df.groupby(['playerID']).sum()
career_batter_df
```

```python
# dropping 'yearID' and 'stint' from career_batter_df
career_batter_df = career_batter_df.drop(columns=['yearID', 'stint']).copy()
career_batter_df
```

```python
# Creating series for Batting Average, On-Base Percentage, and Slugging Percentage
batting_average = career_batter_df['H'] / career_batter_df['AB']
on_base_percent = (career_batter_df['H'] + career_batter_df['BB'] + career_batter_df['HBP']) / (career_batter_df['AB'] + career_batter_df['BB'] + career_batter_df['HBP'] + career_batter_df['SF'])
slg_percent = ((career_batter_df['H'] - (career_batter_df['2B'] + career_batter_df['3B'] + career_batter_df['HR'])) + (career_batter_df['2B'] * 2) + 
(career_batter_df['3B'] * 3) + (career_batter_df['HR'] * 4)) / career_batter_df['AB']
```

```python
# Adding BA, OBP, SLG columns to career_batter_df
career_batter_df['AVG'] = batting_average
career_batter_df['OBP'] = on_base_percent
career_batter_df['SLG'] = slg_percent
career_batter_df
```

```python
# Filling remaining NaN figures as 0
career_batter_df = career_batter_df.fillna(0)
career_batter_df
```

```python
# add inducted Y or N column from HOF_df (changed from 'inner' to 'outer')

hall_batter_df = career_batter_df.merge(HOF_df, how='outer', on='playerID')
hall_batter_df
```

```python
# Dropping unneeded columns

hall_batter_df = hall_batter_df.drop(['yearID', 'votedBy', 'ballots', 'needed', 'votes', 'category', 'needed_note'], axis=1)
hall_batter_df
```

```python
hall_batter_df['inducted'].value_counts()
```

```python
# Dropping duplicate rows of multiple HOF vote receivers, keeping last one for Y votes

hall_batter_df = hall_batter_df.drop_duplicates(subset=['playerID'], keep='last')
hall_batter_df
```

```python
# Filtering to remove players who have never had a MLB at bat

hall_batter_df = hall_batter_df[hall_batter_df.AB > 0]
hall_batter_df
```

```python
# dropping players with under 3000 at bats who have been inducted to the hall of fame (to remove hall of fame pitchers)

hall_batter_df = hall_batter_df.loc[~((hall_batter_df['AB'] < 3000) & (hall_batter_df['inducted'] == 'Y')),:]
hall_batter_df
```

```python
hall_batter_df['inducted'].value_counts()
```

```python
hall_batter_df = hall_batter_df.fillna("N")
hall_batter_df
```

```python
hall_batter_df['inducted'].value_counts()
```

```python
# Setting playerID as index

hall_batter_df = hall_batter_df.set_index('playerID')
hall_batter_df
```

```python
hall_batter_df.to_csv('./Resources/Revised_CSV/hall_batter.csv')
career_batter_df.to_csv('./Resources/Revised_CSV/career_batter.csv')
```

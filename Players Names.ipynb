```python
import pandas as pd
from pathlib import Path


file_path = Path('People.csv')
pitching_df = pd.read_csv(file_path, dtype=object)
pitching_df
```

```python
pitching_df = pitching_df.drop(columns = ['weight','height','bats','throws'])
pitching_df
```

```python
name_df = pd.DataFrame([pitching_df.playerID, pitching_df.nameFirst, pitching_df.nameLast]).transpose()
name_df
```

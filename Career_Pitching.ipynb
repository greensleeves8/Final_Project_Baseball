```python
import pandas as pd
from pathlib import Path


file_path = Path('./Resources/Original_CSV/Pitching.csv')
pitching_df = pd.read_csv(file_path)
pitching_df
```

```python
#merge Hall of Fame into our pitching_df
file_path_HOF = Path('./Resources/Original_CSV/HallOfFame.csv')
HOF_df = pd.read_csv(file_path_HOF)
HOF_df
```

```python
HOF_df = HOF_df.loc[(HOF_df['category'] == 'Player')]
HOF_df
```

```python
# checking for null values
pitching_df.isnull().sum()
```

```python
# dropping SH, SF, GIDP, IBB
pitching_df = pitching_df.drop(columns=['SH', 'SF', 'IBB', 'GIDP', 'yearID' , 'stint']).copy()
pitching_df
```

```python
# Creating new database of players' cumulative career statistics
career_pitching_df = pitching_df.groupby(['playerID']).sum()
career_pitching_df
```

```python
# create IP
innings_pitched = career_pitching_df['IPouts'] / 3.0 
career_pitching_df['IP'] = innings_pitched
```

```python
# creating the correct ERA, win%, Strikeout-to-walk Ratio (K/BB), walks plus hits per inning pitched (WHIP)
win_percentage = career_pitching_df['W'] / (career_pitching_df['W'] + career_pitching_df['L'])
strikout_walk_ratio = career_pitching_df['SO'] / career_pitching_df['BB']
walks_plus_hits_per_ip = career_pitching_df['BB'] + career_pitching_df['H']
ERA_corrected = career_pitching_df['ER'] / career_pitching_df['IP']
```

```python
career_pitching_df['Win %'] = win_percentage
career_pitching_df['(K/BB)'] = strikout_walk_ratio
career_pitching_df['WHIP'] = walks_plus_hits_per_ip / career_pitching_df['IP']
career_pitching_df['ERA_corrected'] = ERA_corrected * 9
career_pitching_df
```

```python
# Filling remaining NaN figures as 0
career_pitching_df = career_pitching_df.fillna(0)
career_pitching_df
```

```python
# filtering career_pitching_df by SO 
# career_pitching_df = career_pitching_df[career_pitching_df.SO>200]
# career_pitching_df
```

```python
# add inducted Y or N column from HOF_df and remove unneeded columns
hall_pitching_df = career_pitching_df.merge(HOF_df, how='outer', on='playerID')
hall_pitching_df = hall_pitching_df.drop(['yearID', 'votedBy', 'ballots', 'needed', 'votes', 'category', 'needed_note', 'BAOpp', 'BK', 'IPouts', 'ERA', 'GF'], axis=1)
hall_pitching_df
```

```python
# view updated columns
hall_pitching_df.columns.tolist()
```

```python
# Dropping duplicate rows of multiple HOF vote receivers, keeping last one for Y votes
hall_pitching_df = hall_pitching_df[hall_pitching_df.SO > 0]
hall_pitching_df = hall_pitching_df.drop_duplicates(subset=['playerID'], keep='last').set_index('playerID')
hall_pitching_df
```

```python
# convert NaN figures to 'N'
hall_pitching_df = hall_pitching_df.fillna("N")
hall_pitching_df
```

```python
# filtering career_pitching_df by SO 
career_pitching_df = career_pitching_df.drop(columns=['IPouts', 'BK', 'GF', 'BAOpp', 'ERA'])
career_pitching_df = career_pitching_df[career_pitching_df.SO>25]
career_pitching_df
```

```python
# Rename columns to comply with Postgres naming conventions
hall_pitching_df.rename(columns={"Win %": "Win Percentage", "(K/BB)": "Strikeout to Walk"}, inplace=True)
career_pitching_df.rename(columns={"Win %": "Win Percentage", "(K/BB)": "Strikeout to Walk"}, inplace=True)
#list(hall_pitching_df.columns)
```

```python
hall_pitching_df.to_csv('./Resources/Revised_CSV/hall_pitching.csv')
career_pitching_df.to_csv('./Resources/Revised_CSV/career_pitching.csv')
```

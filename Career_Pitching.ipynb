```python
import pandas as pd
from pathlib import Path


file_path = Path('./Resources/Pitching.csv')
pitching_df = pd.read_csv(file_path)
pitching_df
```

```python
#merge Hall of Fame into our pitching_df
file_path_HOF = Path('./Resources/HallOfFame.csv')
HOF_df = pd.read_csv(file_path_HOF)
HOF_df
```

```python
# checking for null values
pitching_df.isnull().sum()
```

```python
# dropping SH, SF, GIDP, IBB
pitching_df = pitching_df.drop(columns=['SH', 'SF', 'IBB', 'GIDP', 'yearID' , 'stint']).copy()
pitching_df
```

```python
# Creating new database of players' cumulative career statistics
career_pitching_df = pitching_df.groupby(['playerID']).sum()
career_pitching_df
```

```python
# creating the win%, Strikeout-to-walk Ratio (K/BB), 
win_percentage = career_pitching_df['W'] / career_pitching_df['L']
strikout_walk_ratio = career_pitching_df['SO'] / career_pitching_df['BB']
```

```python
career_pitching_df['Win %'] = win_percentage
career_pitching_df['(K/BB)'] = strikout_walk_ratio
career_pitching_df
```

```python
# Filling remaining NaN figures as 0
career_pitching_df = career_pitching_df.fillna(0)
career_pitching_df
```

```python
# filtering career_pitching_df by SO 
career_pitching_df = career_pitching_df[career_pitching_df.SO>300]
career_pitching_df
```

```python
# change column names to names accepted by Postgres
career_pitching_df = career_pitching_df.rename(columns={'Win %': 'Win Percentage', '(K/BB)': 'K_to_BB'})
```

```python
# add inducted Y or N column from HOF_df and remove unneeded columns
hall_pitching_df = career_pitching_df.merge(HOF_df, how='inner', on='playerID')
hall_pitching_df = hall_pitching_df.drop(['yearID', 'votedBy', 'ballots', 'needed', 'votes', 'category', 'needed_note'], axis=1)
hall_pitching_df
```

```python
# view updated columns
hall_pitching_df.columns.tolist()
```

```python
# Dropping duplicate rows of multiple HOF vote receivers, keeping last one for Y votes
hall_pitching_df = hall_pitching_df.drop_duplicates(subset=['playerID'], keep='last').set_index('playerID')
hall_pitching_df
```

```python
# Rename columns 'Win %' and '(K/BB)' to acceptable name in Postgres
hall_pitching_df = hall_pitching_df.rename(columns={'Win %': 'Win Percentage', '(K/BB)': 'K_to_BB'})
```

```python
# create_engine from the sqlalchemy module
from sqlalchemy import create_engine
```

```python
# Save pitching dataframe to SQL table
hall_pitching_df.to_sql(name='hall_pitching', con=engine, if_exists='replace')
career_pitching_df.to_sql(name='career_pitching', con=engine, if_exists='replace')
```
